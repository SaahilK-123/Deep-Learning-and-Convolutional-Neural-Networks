```python
from ultralytics import YOLO

def train_model():
    print("Loading YOLO26-Small architecture...")
    model = YOLO("yolo26s.pt")

    yaml_path = "data.yaml"

    print("Initiating transfer learning on merged dataset...")

    results = model.train(
    data='./dataset/data.yaml',
    epochs=200,
    imgsz=640,
    batch=8,
    project='runs/detect',
    name='phase1_yolo26s_model',
    plots=True,
    patience=30,
    freeze=10,
    fliplr=0.5,
    degrees=15.0,
    hsv_v=0.4,
    hsv_s=0.7,
)

    print("Training complete! Weights saved in 'runs\detect\phase1_yolo26s_results\weights\best.pt'")

if __name__ == "__main__":
    train_model()
```
